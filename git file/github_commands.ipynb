| Situation                                  | Use `git init`?                                    |
| ------------------------------------------ | -------------------------------------------------- |
| Starting a brand new project               | ✅ Yes                                              |
| You downloaded a folder that is not a repo | ✅ Yes                                              |
| You already cloned a repository            | ❌ No (because `git clone` already includes `.git`) |


| Command          | New Files      | Modified Files | Deleted Files  | Notes                             |
| ---------------- | -------------- | -------------- | -------------- | --------------------------------- |
| `git add .`      | Yes            | Yes            | Some           | Adds everything in current folder |
| `git add -A`     | Yes            | Yes            | Yes            | Best for “add everything”         |
| `git add -u`     | No             | Yes            | Yes            | Only updates, no new files        |
| `git add <file>` | Only that file | Only that file | Only that file | Manual control                    |

## To change the user name, email and repository:
|T0 remove the user name and mail:
---------------------------------
*git config --global --unset user.name*
*git config --global --unset user.email*
it will automatically remove the user namer and mail
to check: *git config --list*
if the use name and user mail is not appearing than it was successfully deleted

|To add new one:
----------------
git config --*global user.name "Your New Name"*
git config --*global user.email "yournewemail@example.com"*

|For origin replacemnt:
-----------------------
To remove the old origin: git remote remove origin
git remote add origin https://github.com/USERNAME/NEW-REPO.git (username: im-krishh, new repo: project01)
Or we can directly change the existing origin by: git remote set-url origin https://github.com/USERNAME/NEW-REPO.git
